```python
df[['Source', 'Medium']] = df['Source / Medium'].str.split('/', expand=True)
```

```python
df.Medium.unique()
```

```python
def handle_medium(value):
    paid = [' cpc', ' cpm', ' ppc', ' paid', ' affiliate']
    if value == ' (none)':
        value =  'Direct'
    elif value in paid:
        value = 'Paid'
    elif value == ' organic':
        value =  'Organic'
    elif value == ' referral':
        value =  'Referral'
    elif value == ' (not set)':
        value = 'Others'
    return value
```

```python
df['Medium'] = df['Medium'].apply(handle_medium)
```

```python
df.Medium.unique()
```

```python
pivot_table = pd.pivot_table(df, values='Revenue', index='Month of Year', columns='Medium', aggfunc='sum')
```

```python
grouped_df = df.groupby(['Month of Year', 'Medium'])['Revenue'].sum().reset_index()
```

```python
for month in grouped_df['Month of Year'].unique():
    # Filter the data for the specific month
    month_data = grouped_df[grouped_df['Month of Year'] == month]
    
    # Create a pie chart
    plt.figure(figsize=(4, 4))
    plt.pie(month_data['Revenue'], labels=month_data['Medium'], autopct='%1.1f%%',startangle=90, pctdistance =1)
    plt.title(f'Revenue from each Medium {month}')
    plt.legend(loc = 'lower left')
    plt.show()
```
